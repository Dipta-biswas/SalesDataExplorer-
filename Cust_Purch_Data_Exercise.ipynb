**6. Two customers have the same phone number, can you find those customers**?

```python
# Code here please
df['phone'].value_counts().head(2)
```

```python
# lets find what phone number is twice!
```

```python
# Code here please
df["full_name"] = df["prefix"] + " " + df["first"] + " " + df["last"]

# Find duplicate phone numbers
duplicate_phones = df[df.duplicated("phone", keep=False)]

# Display the customers with duplicate phone numbers
print(duplicate_phones[["full_name", "phone"]])
```

**10. What is the max, min and average spending?**

```python
# Code here please

# Calculate max, min, and average spending
max_spending = df["price(CAD)"].max()
min_spending = df["price(CAD)"].min()
avg_spending = df["price(CAD)"].mean()

print(f"Max spending: {max_spending} CAD")
print(f"Min spending: {min_spending} CAD")
print(f"Average spending: {avg_spending:} CAD")
```

**18. Can you tell the top 5 most popular email providers? (e.g. gmail.com, yahoo.com, etc...)**
